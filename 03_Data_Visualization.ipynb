## 1. Introduction to Matplotlib

#### Example Usage:

Now, let's go ahead and create more fancier plots using matplotlib!!!

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
# Generate sample data
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
temp_max = np.random.normal(loc=25, scale=5, size=len(dates)) + 10 * np.sin(np.arange(len(dates)) * 2 * np.pi / 365)
temp_min = temp_max - np.random.uniform(5, 10, size=len(dates))

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(12, 6))

# Plot the data
ax.plot(dates, temp_max, label='Max Temperature')
ax.plot(dates, temp_min, label='Min Temperature')

# Customize the plot
ax.set_title('Daily Temperature Variation (2023)', fontsize=16)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Temperature (Â°C)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)
ax.legend(fontsize=10)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

In this example, we've used Matplotlib's object-oriented interface to create a more complex plot with two lines representing maximum and minimum temperatures. Key points to note:
 
- We use `fig, ax = plt.subplots()` to create a figure and axis object.
- We plot directly on the axis object (`ax.plot()`).
- We customize various aspects of the plot using axis methods.
- The `tight_layout()` function adjusts the plot to fit within the figure area.

### Subplots in Matplotlib
Matplotlib allows us to create multiple plots in a single figure using subplots.

```python
# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2, figsize=(12, 10))

# Flatten the 2D array of axes for easier indexing
axs = axs.flatten()

# Plot 1: Line plot
axs[0].plot(dates, temp_max)
axs[0].set_title('Max Temperature')
axs[0].set_xlabel('Date')
axs[0].set_ylabel('Temperature (Â°C)')

# Plot 2: Scatter plot
axs[1].scatter(temp_max, temp_min, alpha=0.5)
axs[1].set_title('Max vs Min Temperature')
axs[1].set_xlabel('Max Temperature (Â°C)')
axs[1].set_ylabel('Min Temperature (Â°C)')

# Plot 3: Histogram
axs[2].hist(temp_max, bins=30, edgecolor='black')
axs[2].set_title('Distribution of Max Temperature')
axs[2].set_xlabel('Temperature (Â°C)')
axs[2].set_ylabel('Frequency')

# Plot 4: Bar plot
monthly_avg = pd.DataFrame({'temp': temp_max}, index=dates).resample('M').mean()
axs[3].bar(monthly_avg.index.strftime('%b'), monthly_avg['temp'])
axs[3].set_title('Monthly Average Max Temperature')
axs[3].set_xlabel('Month')
axs[3].set_ylabel('Temperature (Â°C)')
plt.xticks(rotation=45)

# Adjust layout and display the plot
plt.tight_layout()
plt.show()
```

This example demonstrates how to create and customize multiple subplots in a single figure. We've used different types of plots (line, scatter, histogram, and bar) to show various ways of visualizing the same dataset.

#### Example Usage:

```python
import seaborn as sns
import pandas as pd

# Sample dataset
data = pd.DataFrame({
    'Category': ['A', 'B', 'C', 'D'],
    'Values': [10, 20, 15, 25]
})

# Create a bar plot
sns.barplot(x='Category', y='Values', data=data)
```

```python
# Importing Seaborn
import seaborn as sns

# Example: Create a boxplot of temperatures
temps = [22.5, 25.3, 20.8, 18.6, 30.0]
sns.boxplot(temps)
plt.title('Temperature Distribution')
plt.show()
```

```python
# Generate a more complex dataset
np.random.seed(42)
n_samples = 1000
data = pd.DataFrame({
    'Date': pd.date_range(start='2023-01-01', end='2023-12-31', periods=n_samples),
    'Temperature': np.random.normal(loc=20, scale=5, size=n_samples),
    'Humidity': np.random.uniform(30, 100, size=n_samples),
    'Pressure': np.random.normal(loc=1013, scale=5, size=n_samples),
    'WindSpeed': np.random.exponential(scale=5, size=n_samples),
    'Precipitation': np.random.exponential(scale=5, size=n_samples)
})
data['Season'] = data['Date'].dt.month.map({1:'Winter', 2:'Winter', 3:'Spring', 4:'Spring', 5:'Spring', 
                                            6:'Summer', 7:'Summer', 8:'Summer', 9:'Fall', 10:'Fall', 
                                            11:'Fall', 12:'Winter'})

# Create a pair plot
sns.pairplot(data, vars=['Temperature', 'Humidity', 'Pressure', 'WindSpeed'], 
             hue='Season', palette='viridis')
plt.suptitle('Relationships Between Weather Variables', y=1.02)
plt.show()
```

### Code Explanation: Creating a Complex Weather Dataset and Pair Plot with Seaborn

In this code, we generate a complex dataset simulating various weather attributes across 1000 samples and visualize the relationships between these variables using Seaborn's `pairplot`. This is a great way to explore relationships between multiple continuous variables and group them by categorical labels (in this case, seasons).

#### Step-by-step Breakdown:

1. **Generating a Complex Dataset:**
   - `np.random.seed(42)` ensures reproducibility for random number generation.
   - `n_samples = 1000` specifies that the dataset will contain 1000 samples.
   - `data = pd.DataFrame({...})` creates a `DataFrame` with the following variables:
     - `Date`: Randomly distributed dates throughout the year 2023.
     - `Temperature`: Normally distributed temperature values with a mean of 20Â°C and a standard deviation of 5Â°C.
     - `Humidity`: Uniformly distributed humidity values between 30% and 100%.
     - `Pressure`: Normally distributed atmospheric pressure values centered around 1013 hPa with a standard deviation of 5 hPa.
     - `WindSpeed` and `Precipitation`: Exponentially distributed values for wind speed and precipitation, simulating their natural distribution.
   - The `Season` column is added based on the month extracted from the `Date` column, mapping months to their respective seasons.

2. **Creating a Pair Plot:**
   - `sns.pairplot(data, vars=['Temperature', 'Humidity', 'Pressure', 'WindSpeed'], hue='Season', palette='viridis')` generates a pair plot for the selected weather variables (`Temperature`, `Humidity`, `Pressure`, `WindSpeed`), and colors the data points by season using the "viridis" color palette.
     - **Pair plots** show pairwise relationships between variables in the form of scatter plots and histograms, making it easy to explore interactions between multiple variables.
     - The `hue='Season'` parameter distinguishes data points based on the season, adding context to the relationships.
   
3. **Plot Title and Display:**
   - `plt.suptitle('Relationships Between Weather Variables', y=1.02)` adds a title to the figure, placed slightly above the plot for better alignment.
   - `plt.show()` displays the resulting visualization, which provides insight into how different weather variables are related across different seasons.

This pair plot helps visually explore the correlations and distributions between multiple weather factors while highlighting seasonal variations.


```python
# Create a more complex plot combining multiple Seaborn functions
fig, axs = plt.subplots(2, 2, figsize=(15, 15))

# Plot 1: Scatter plot with regression line
sns.regplot(data=data, x='Temperature', y='Humidity', ax=axs[0, 0])
axs[0, 0].set_title('Temperature vs Humidity with Regression Line')

# Plot 2: Box plot
sns.boxplot(data=data, x='Season', y='Temperature', ax=axs[0, 1])
axs[0, 1].set_title('Temperature Distribution by Season')

# Plot 3: Violin plot
sns.violinplot(data=data, x='Season', y='WindSpeed', ax=axs[1, 0])
axs[1, 0].set_title('Wind Speed Distribution by Season')

# Plot 4: Heatmap of correlation matrix
corr = data[['Temperature', 'Humidity', 'Pressure', 'WindSpeed', 'Precipitation']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axs[1, 1])
axs[1, 1].set_title('Correlation Heatmap')

plt.tight_layout()
plt.show()
```

This example showcases various Seaborn plot types:
- `regplot`: Scatter plot with a regression line, useful for visualizing linear relationships.
- `boxplot`: Shows the distribution of a variable across categories.
- `violinplot`: Similar to a box plot, but shows the full distribution.
- `heatmap`: Useful for visualizing correlation matrices or other 2D data.

## 3. Introduction to Cartopy

**Cartopy** is a powerful Python library designed for geospatial data visualization and mapping. It provides tools for creating maps and handling spatial data, making it highly suitable for projects that involve geographic data, such as weather patterns, global distributions, or tracking events across the globe.

#### Key Features:
- **Map Projections:** Cartopy supports a wide variety of map projections, allowing you to represent the Earth's surface in different ways.
- **Geospatial Data Handling:** It can handle and plot geospatial data from various sources, such as shapefiles, raster images, and more.
- **Integration with Matplotlib:** Cartopy is built on top of Matplotlib, which makes it easy to combine geographical data with Matplotlib's rich plotting features.
- **Coastlines and Boundaries:** The library includes utilities for adding coastlines, country borders, rivers, and other natural boundaries to your maps.
- **Efficient for Large Datasets:** Cartopy efficiently handles large geospatial datasets, making it an ideal choice for visualizing complex data across the globe.

#### Example Usage:
Hereâ€™s an example of how you can create a simple map with Cartopy:


```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
import cartopy.feature as cfeature

# Create a map using the PlateCarree projection
ax = plt.axes(projection=ccrs.PlateCarree())
ax.coastlines()

plt.title('Simple World Map with Cartopy')
plt.show()
```

```python
# Generate sample global temperature data
def generate_temp_data(lons, lats):
    lon_grid, lat_grid = np.meshgrid(lons, lats)
    return 15 + 10 * np.cos(np.deg2rad(lat_grid)) + np.random.randn(*lon_grid.shape)

lons = np.linspace(-180, 180, 72)
lats = np.linspace(-90, 90, 36)
lon_grid, lat_grid = np.meshgrid(lons, lats)
temp_data = generate_temp_data(lons, lats)

# Create a global map with temperature data
fig = plt.figure(figsize=(15, 10))
ax = plt.axes(projection=ccrs.Robinson())

# Add map features
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.OCEAN, zorder=0)
ax.add_feature(cfeature.LAND, zorder=0, edgecolor='black')

# Plot the temperature data
cs = ax.contourf(lon_grid, lat_grid, temp_data, transform=ccrs.PlateCarree(),
                 cmap='RdYlBu_r', extend='both')

# Add colorbar and title
cbar = plt.colorbar(cs, orientation='horizontal', pad=0.05, aspect=50)
cbar.set_label('Temperature (Â°C)')

plt.title('Global Temperature Distribution', fontsize=16)
plt.show()
```

### Code Explanation: Global Temperature Visualization with Cartopy

In this code, we generate a synthetic dataset representing global temperatures and plot it on a world map using **Cartopy**. The map includes various geographical features such as coastlines and borders, and we use contour plotting to visualize the temperature distribution.

#### Step-by-step Breakdown:

1. **Generate Sample Temperature Data:**
   - `generate_temp_data(lons, lats)` is a function that creates synthetic temperature data for given longitude (`lons`) and latitude (`lats`) arrays.
   - The function models temperature as a cosine function of latitude, with added random noise for variability.
   - `lons = np.linspace(-180, 180, 72)` and `lats = np.linspace(-90, 90, 36)` define the longitude and latitude ranges for the temperature grid, covering the entire globe.

2. **Create a Global Map with Cartopy:**
   - `ax = plt.axes(projection=ccrs.Robinson())` sets up a map using the Robinson projection, which provides an appealing, distortion-reduced view of the entire world.
   - We add geographic features like coastlines (`ax.add_feature(cfeature.COASTLINE)`), country borders (`ax.add_feature(cfeature.BORDERS)`), and land/ocean areas.

3. **Plot the Temperature Data:**
   - `ax.contourf(lon_grid, lat_grid, temp_data, transform=ccrs.PlateCarree(), cmap='RdYlBu_r')` plots the temperature data using a filled contour plot (`contourf`). The `PlateCarree` projection is used to plot the data in latitude-longitude coordinates, while the `RdYlBu_r` colormap is used to visualize temperature gradients (from red for warm to blue for cold).

4. **Add Colorbar and Title:**
   - `plt.colorbar(cs, orientation='horizontal', pad=0.05, aspect=50)` adds a horizontal colorbar beneath the plot to represent temperature values. The `extend='both'` argument ensures the colorbar represents data beyond the colormap range.
   - `plt.title('Global Temperature Distribution', fontsize=16)` adds a descriptive title to the map.

5. **Display the Plot:**
   - `plt.show()` renders the global temperature map, giving us a visual representation of temperature variations across the globe.

This plot effectively combines Cartopy's geographic capabilities with Matplotlib's contour plotting to display global temperature data.


```python
# Create a map focused on a specific region
fig = plt.figure(figsize=(15, 10))
ax = plt.axes(projection=ccrs.LambertConformal(central_longitude=-100, central_latitude=35))

# Set map extent (North America)
ax.set_extent([-130, -60, 20, 55], crs=ccrs.PlateCarree())

# Add map features
ax.add_feature(cfeature.COASTLINE)
ax.add_feature(cfeature.BORDERS)
ax.add_feature(cfeature.STATES)
ax.add_feature(cfeature.OCEAN)
ax.add_feature(cfeature.LAND)

# Plot some sample weather station data
stations = {
    'New York': (-74.006, 40.7128),
    'Los Angeles': (-118.2437, 34.0522),
    'Chicago': (-87.6298, 41.8781),
    'Houston': (-95.3698, 29.7604),
    'Phoenix': (-112.0740, 33.4484)
}

for city, (lon, lat) in stations.items():
    ax.plot(lon, lat, 'ro', transform=ccrs.PlateCarree(), markersize=10)
    ax.text(lon, lat, city, transform=ccrs.PlateCarree(), fontsize=8,
            horizontalalignment='right', verticalalignment='bottom')

plt.title('Weather Stations in North America', fontsize=16)
plt.show()
```

### Code Explanation: Creating a Map Focused on North America with Weather Stations

     - `cfeature.OCEAN` and `cfeature.LAND` distinguish ocean and land areas on the map.

3. **Plotting Weather Stations:**
   - `stations` is a dictionary containing the longitude and latitude coordinates of key cities.

- Exploratory Data Analysis: Use Matplotlib and Seaborn to create quick plots and understand your data.
- Statistical Visualization: Use Seaborn for more advanced statistical graphics.
- Geospatial Visualization: Use Cartopy for mapping and spatial data visualization.
- Customization: Fine-tune your plots using Matplotlib's lower-level functions for publication-quality figures.
 
**Note:** Remember to always consider your audience and choose the most appropriate visualization for your data and message. With practice, you'll be able to create informative and visually appealing graphics to communicate your research effectively.
